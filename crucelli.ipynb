# Pipeline Crucelli

Esegue in ordine:
1) tutti gli script in `src/data_update/`,
2) `src/queries/regola_1.py`,
3) `src/queries/betting.py`.

Poi mostra alcune tabelle di output se presenti.

```python
from __future__ import annotations

import runpy
from pathlib import Path

ROOT = Path.cwd()
DATA_UPDATE_DIR = ROOT / "src" / "data_update"
QUERIES_DIR = ROOT / "src" / "queries"

def run_script(path: Path) -> None:
    print(f"\n--- Running: {path} ---")
    runpy.run_path(str(path), run_name="__main__")

# 1) Run all data_update scripts in alphabetical order
update_scripts = sorted(DATA_UPDATE_DIR.glob("*.py"))
for script in update_scripts:
    run_script(script)

# 2) Run regola_1.py
run_script(QUERIES_DIR / "regola_1.py")

# 3) Run betting.py
run_script(QUERIES_DIR / "betting.py")
```
